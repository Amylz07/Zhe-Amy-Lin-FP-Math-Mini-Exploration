```python
# iterated values for a single iterative sequence

from tabulate import tabulate

x = 1
n = 20
K = 2
i = 0
x_list = []
x_list.append(x)
while i<n:
    x = K*x+5
    x_list.append(x)
    i += 1
    
def iteration_times(it_start, it_end):
    return list(range(it_start, it_end+1))
it_start = 0
it_end = n
it_list = iteration_times(it_start, it_end)

print(tabulate({'n': it_list, 'x_n': x_list}))
```

```python
# scatter plots for linear iterative sequences with changing K

from tabulate import tabulate

K_init = -2
K = K_init

while K < 2:
    x = 1
    n = 25
    i = 0
    x_list = []
    x_list.append(x)
    while i<n:
        x = K*x+5
        x_list.append(x)
        i += 1

    def iteration_times(it_start, it_end):
        return list(range(it_start, it_end+1))
    it_start = 0
    it_end = n
    it_list = iteration_times(it_start, it_end)

    import matplotlib.pyplot as plt
    # plt.figure(figsize=(40, 30),dpi=400)
    fig,ax=plt.subplots(dpi=800)
    ax.scatter(it_list, x_list)
    ax.set_xlabel("number of iterations")
    ax.set_ylabel("x")
    x_ = "$x_{n-1}$"
    ax.set_title("$x_n$={}{}+5".format(round(K,2), x_))
    
    K += 0.1
```

```python
# scatter plots for quadratic iterative sequence with changing K

from tabulate import tabulate

K_init = -2
K = K_init

while K < 4:
    x = 0.2
    n = 100
    i = 0
    x_list = []
    x_list.append(x)
    while i<n:
        x = K*x*(1-x)
        x_list.append(x)
        i += 1

    def iteration_times(it_start, it_end):
        return list(range(it_start, it_end+1))
    it_start = 0
    it_end = n
    it_list = iteration_times(it_start, it_end)

    import matplotlib.pyplot as plt
    fig,ax=plt.subplots(dpi=800)
    ax.scatter(it_list, x_list)
    ax.set_xlabel("number of iterations")
    ax.set_ylabel("x")
    x_ = "$x_{n-1}$"
    ax.set_title("$x_n$={}{}(1-{})".format(round(K,2), x_, x_))
    
    K += 0.1
```

```python
# staircase diagram for linear iterative sequence

import matplotlib.pyplot as plt
from numpy import linspace
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 800

def cobweb (f, x0, N, it, a=-2.5, b=15):
    t = linspace(a, b, N)
    plt.plot(t, f(t), 'r', linewidth=1)
    
    plt.plot(t, t, 'r:', linewidth=1)
    
    x, y = x0, f(x0)
    i = 0
    while i < it:
        plt.plot([x, y], [y, y], 'b', linewidth=0.7)
        plt.plot([y, y], [y, f(y)], 'b', linewidth=0.7)
        plt.xlabel('x')
        plt.ylabel('y')
        x_ = '$x_{n-1}$'
        plt.title('$x_n$={}{}+5, $x_0$={}'.format(K, x_, x0))
        x, y = y, f(y)
        i += 1
        
K=0.5
cobweb(lambda x: K*x+5, 1, 10000, 8)
```

```python
# cobweb diagram for quadratic iterative sequence

import matplotlib.pyplot as plt
from numpy import linspace
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 800

def cobweb (f, x0, N, it, a=-0.1, b=1.1):
    t = linspace(a, b, N)
    plt.plot(t, f(t), 'r', linewidth=1)
    
    plt.plot(t, t, 'r:', linewidth=1)
    
    x, y = x0, f(x0)
    i = 0
    while i < it:
        plt.plot([x, y], [y, y], 'b', linewidth=0.7)
        plt.plot([y, y], [y, f(y)], 'b', linewidth=0.7)
        plt.xlabel('x')
        plt.ylabel('y')
        x_ = '$x_{n-1}$'
        plt.title('$x_n$={}{}(1-{}), $x_0$={}'.format(K, x_, x_, x0))
        x, y = y, f(y)
        i += 1

K=3.8
cobweb(lambda x: K*x*(1-x), 0.2, 10000, 100)
```
